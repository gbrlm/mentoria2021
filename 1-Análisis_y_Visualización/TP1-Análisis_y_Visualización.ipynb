### Algunos ejemplos

Veamos algunas variables como ejemplo.

Tomemos la pregunta M0P0 "Frecuencia con que se han sufrido diferentes comportamientos de la pareja actual: violencia psicológica (control social)". Esta es una pregunta de múltiple opción. Entonces las columnas que nos interesan son las que tienen por nombre: 

M1P1_0_1,M1P1_0_2,...,M1P1_0_11.

```python
cols_M1P1 = [col for col in dataset.columns if 'M1P1_0_' in col]
```

Así podemos trabajar con un dataset más chico seleccionando estas columnas

```python
dataset_M1P1 = dataset[cols_M1P1]
display(dataset_M1P1)
```

De aquí, un primer análisis que podemos hacer es contar la cantidad de respuestas

```python
ans_M1P1 = dataset_M1P1.apply(pd.Series.value_counts, axis=0)
ans_M1P1
```
